```python
def replace_position(df, lib, con, etc):
    """
    lib : 진보당
    con : 보수당
    etc : 나머지당
    """
    df["lib"] = df.loc[:, lib].sum(axis=1)
    df["con"] = df.loc[:, con].sum(axis=1)
    df["etc"] = df.loc[:, etc].sum(axis=1)
    df.drop(columns=(lib + con + etc), inplace=True)
    return df
```

```python
arg_list = list()
partial_df_list = list()

for dir_path in dir_paths:
    file_paths = glob(os.path.join(dir_path, "*.xlsx"))
    for file_path in file_paths:
        raw_df = pd.read_excel(file_path, header=[2, 3, 4, 5])
        df = raw_df.copy()
        #
        theme = df.columns.get_level_values(0)[0]
        _, loc_1, theme, loc_2 = theme.replace("][", ",").strip("[]").split(",")

        #
        df = df.droplevel(level=0, axis=1)
        #
        replace_unnamed = lambda x: None if x.startswith("Unnamed") else x
        replace_escape = lambda x: x.replace("\n", "_")
        replacer = lambda x: replace_unnamed(replace_escape(x))
        flatten_columns = ["_".join(filter(None, map(replacer, col))) for col in df.columns]
        df.columns = flatten_columns

        #
        df = df.drop(columns=["후보자별 득표수__"])

        #
        df = df[df["투표구명"] == "소계"]

        #
        rename_dict = {"읍면동명": "loc_3", "무효_투표수": "무효투표수"}
        df = df.rename(columns=rename_dict)

        df["loc_1"] = loc_1
        df["loc_2"] = loc_2
        df["theme"] = theme

        #
        votes = list(filter(lambda x: x.startswith("후보자별 득표수_"), df.columns))
        df.loc[:, votes] = df.loc[:, votes].astype(int)
        lib = list(filter(lambda x: x.startswith("후보자별 득표수_더불어민주당"), df.columns))
        con = list(filter(lambda x: x.startswith("후보자별 득표수_미래통합당"), df.columns))
        etc = list(set(votes) - (set(lib) | set(con)))
        df = replace_position(df, lib, con, etc)

        #
        argument = loc_1 + "_" + loc_2 + "_" + theme
        arg_list.append(argument)
        partial_df_list.append(df)
```

```python
국회의원_21 = pd.concat(partial_df_list, axis=0)
```

```python
국회의원_21.rename(columns={"theme": "선거구명"}, inplace=True)
```
